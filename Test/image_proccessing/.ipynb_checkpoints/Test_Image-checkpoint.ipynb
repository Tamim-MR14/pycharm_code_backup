```python
import numpy as np
import pandas as pd
from PIL import Image
from PIL import *
from PIL import ImageFilter
```

```python
import matplotlib.pyplot as plt 
image = Image.open("1-0.jpg")
im = image.convert('L')
im_array=np.array(im)
plt.imshow(im_array,cmap='gray')
```

```python
p = image.convert("L").filter(ImageFilter.GaussianBlur(radius = 2))
p = p.resize((4,4),Image.ANTIALIAS)
P_array=np.array(p).flatten()
q=image.convert("L").filter(ImageFilter.EDGE_ENHANCE())
q= q.resize((4,4),Image.ANTIALIAS)
Q_array=np.array(q).flatten()
P_array=np.concatenate([P_array,Q_array])
P_array
```

```python
import os
dirname='./trainingSample/'
folders=os.listdir(dirname)
data=[]
label=[]

# os.path.join(dirname,'0')
for folder in folders:
    folder_path=os.path.join(dirname,folder)
    for img in os.listdir(folder_path):
        img_path=os.path.join(folder_path,img)
        im=Image.open(img_path)
        p = im.filter(ImageFilter.CONTOUR())
        p = p.resize((4,4),Image.ANTIALIAS)
        P_array=np.array(p).flatten()
        q=im.filter(ImageFilter.EDGE_ENHANCE())
        q= q.resize((4,4),Image.ANTIALIAS)
        Q_array=np.array(q).flatten()
        P_array=np.concatenate([P_array,Q_array])
        data+=[P_array]
        label+=[int(folder)]
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
data=np.array(data)
label=np.array(label)

X_train,X_test,Y_train,Y_test=train_test_split(data,label,test_size=0.2)

clf =MLPClassifier(hidden_layer_sizes=(700,500,400),activation='identity',solver='lbfgs')

clf = clf.fit(X_train, Y_train)
tr_total=clf.score(X_train,Y_train)
score=clf.score(X_test,Y_test)


print(score)
```
